# Introduction

There is a well-known saying by the German writer Johann Wolfgang von Goethe:

> “Tell me with whom you consort and I will tell you who you are…”

Goethe’s idea is that the people you spend time with every day are a reflection of who you are as a person. Would you agree?

Now, imagine extending this idea from people to words. Linguist John Rupert Firth did exactly that and formulated his own principle:

> “You shall know a word by the company it keeps.”

This idea — that a word’s meaning can be understood by its context, or the words that surround it — is the foundation of **word embeddings**.

A **word embedding** is a representation of a word as a numeric vector. This allows us to:

* compare how words are used,
* measure how similar or different words are,
* identify words that occur in similar contexts.

Common applications of word embeddings include:

* entity recognition in chatbots
* sentiment analysis
* syntax parsing

# Vectors

Vectors appear in many fields, but at their core, they act as **containers of information**. The amount of information a vector holds depends on its **dimension**.

## 1-Dimensional Vectors

A 1-dimensional vector stores a single number. For instance, if we want to represent the **length of a word**, we could map words like this:

* `"cat"` → `[3]`
* `"scrabble"` → `[8]`
* `"antidisestablishmentarianism"` → `[28]`

By plotting these values on a **number line**, we can easily see that `"cat"` is much shorter than `"scrabble"`, which is far shorter than `"antidisestablishmentarianism"`.




## 2-Dimensional Vectors

Suppose we also want to include the **number of vowels** in each word. We introduce a 2-dimensional vector:

* The first value = word length
* The second value = number of vowels

So:

* `"cat"` → `[3, 1]`
* `"scrabble"` → `[8, 2]`
* `"antidisestablishmentarianism"` → `[28, 11]`

These vectors can be plotted on a **2D grid**, where:

* **x-axis** = number of letters
* **y-axis** = number of vowels




On the grid, `"cat"` and `"scrabble"` appear closer to each other than they do to `"antidisestablishmentarianism"` — meaning their characteristics are more similar.

## Higher Dimensions

Although visualizing vectors above 2 dimensions becomes difficult, we can still create vectors with **any number of dimensions** — even **1,000!**
Vectors **summarize information using numbers**, allowing comparisons between objects numerically. This idea forms the foundation of **word embeddings**, where words are represented as vectors.

## Example in Python (using NumPy)

To create a vector of odd numbers from 1 to 9:

```python

```

Vectors are a powerful way to convert abstract concepts into numerical form — enabling analysis, comparison, and learning from data.

# Distance

In word embeddings, **distance** is the key concept that allows us to compare words numerically. Distance tells us **how similar or different** two vectors are. Several metrics can measure this, and here we explore three common ones.

---

## 1. Manhattan Distance (City Block Distance)

This distance is the **sum of absolute differences** across each vector dimension:

$$
\text{Manhattan distance} = |1 - 2| + |2 - 4| + |3 - 6| = 1 + 2 + 3 = 6
$$

In Python, we can compute it using SciPy’s `cityblock()`:

```python
manhattan_d = cityblock(vector_a, vector_b)  # 6
```

---

## 2. Euclidean Distance (Straight-Line Distance)

This is the familiar distance formula from geometry — **square root of the sum of squared differences**:

$$
\text{Euclidean distance} = \sqrt{(1 - 2)^2 + (2 - 4)^2 + (3 - 6)^2} = \sqrt{14} \approx 3.74
$$

Using SciPy:

```python
euclidean_d = euclidean(vector_a, vector_b)  # 3.74 (approx)
```

---

## 3. Cosine Distance

This metric compares **the angle between two vectors**, not their magnitude.

* Cosine distance = **0** → vectors point in the same direction (similar)
* Cosine distance = **1** → vectors are opposite (very different)

It can be computed as:

```python
cosine_d = cosine(vector_a, vector_b)  # 0.0
```

---

## Why Cosine Distance Is Preferred for Word Embeddings

Word embeddings often have **high-dimensional vectors** (e.g., 96 or 300 dimensions). In those cases:

* Manhattan and Euclidean distances become very large and harder to interpret
* Cosine distance focuses only on **direction**, which captures similarity more reliably

---

## Example Code

```python
from scipy.spatial.distance import cityblock, euclidean, cosine
import numpy as np

vector_a = np.array([1, 2, 3])
vector_b = np.array([2, 4, 6])

manhattan_d = cityblock(vector_a, vector_b)
euclidean_d = euclidean(vector_a, vector_b)
cosine_d = cosine(vector_a, vector_b)
```

Distance is the foundation for comparing words numerically — and it plays a central role in how word embeddings reveal meaning.


````python
# Step 1: Import libraries and load spaCy model
import numpy as np
from scipy.spatial.distance import cityblock, euclidean, cosine
import spacy
nlp = spacy.load("en_core_web_sm")

# Step 2: Define the word embedding vectors
happy_vec = nlp('happy').vector
sad_vec = nlp('sad').vector
angry_vec = nlp('angry').vector

# Step 3: Compute and print Manhattan distances
man_happy_sad = cityblock(happy_vec, sad_vec)
man_sad_angry = cityblock(sad_vec, angry_vec)
print(man_happy_sad)
print(man_sad_angry)

# Step 4: Compute and print Euclidean distances
euc_happy_sad = euclidean(happy_vec, sad_vec)
euc_sad_angry = euclidean(sad_vec, angry_vec)
print(euc_happy_sad)
print(euc_sad_angry)

# Step 5: Compute and print Cosine distances
cos_happy_sad = cosine(happy_vec, sad_vec)
cos_sad_angry = cosine(sad_vec, angry_vec)
print(cos_happy_sad)
print(cos_sad_angry)
````

# Word Embeddings Are All About Distance

The foundation of word embeddings comes from the **distributional hypothesis**, which states that *words that appear in similar contexts tend to have similar meanings*. Word embeddings turn this idea into mathematics by mapping such words to nearby points in a **vector space**.

## Meaning Through Comparison

The specific numeric values inside a word’s embedding do **not** carry direct meaning. Instead, meaning emerges from **relationships**:

* Words used in similar contexts → **small cosine distance**
* Words used in very different contexts → **large cosine distance**

In other words, embeddings matter because they let us **compare** words. The value of a single number in the vector is not interpretable on its own — but the **relative distances** between vectors reveal patterns in language.

## Latent Information

Although the numbers inside the vectors aren't meant to be read directly, they encode **latent linguistic information** learned from large text corpora. This allows models to capture:

* similarity
* analogy
* semantic relationships

By turning words into vectors, we make it possible for computers to analyze meaning through **distance**, unlocking a powerful tool in natural language processing.

# Word2vec

How do we obtain the numeric values that make up a word embedding? And how do we ensure that words used in similar contexts end up with vectors close together, while unrelated words are placed far apart?

This is where **word2vec** comes in. Word2vec is a statistical learning algorithm that creates word embeddings directly from a text corpus. It does this using one of two model architectures:

---

## 1. Continuous Bag-of-Words (CBOW)

CBOW processes the training corpus in order and tries to **predict each target word using the surrounding context words**, treated as a bag-of-words (ignoring order).

For example, given the context window:

> “The squids ___ out of the suitcase”

CBOW tries to predict the missing word using the context around it.
Since the model ignores word order, training is efficient — but sometimes loses nuance.

---

## 2. Continuous Skip-Grams

Skip-grams work in the opposite direction: the model uses the **current word to predict surrounding context words**, while considering the order of the words.

For example, in the sentence:

> “The squids jump out of the suitcase”

The **1-skip-2-grams** include all standard bigrams plus sequences like:

* (The, jump)
* (squids, out)
* (jump, of)
* (out, the)
* (of, suitcase)

Skip-grams incorporate ordering and can capture long-range dependencies, producing **higher-quality embeddings** — though training is slower than CBOW.

---

## The Neural Network Behind word2vec

Both CBOW and skip-grams feed into a shallow **two-layer neural network**.
The hidden layer weights learned during training become the word embeddings. These weights position words so that:

* similar contexts → **nearby vectors**
* different contexts → **distant vectors**

This process ensures that context-driven meaning becomes encoded in the vector space.

---

Both methods, CBOW and skip-grams, provide powerful ways to learn semantic relationships directly from raw text.


````python
# Step 1: Import CountVectorizer and print the original sentence
from sklearn.feature_extraction.text import CountVectorizer

sentence = "It was the best of times, it was the worst of times."
print(sentence)

# Step 2: Preprocess the sentence into a clean list of words
sentence_lst = [word.lower().strip(".") for word in sentence.split()]

# Step 3: Set context_length to 3 as required
context_length = 2

# Step 4: Define the CBOW function (bag-of-words, order ignored)
def get_cbows(sentence_lst, context_length):
    cbows = []
    for i, val in enumerate(sentence_lst):
        if i < context_length:
            pass
        elif i < len(sentence_lst) - context_length:
            context = sentence_lst[i-context_length:i] + sentence_lst[i+1:i+context_length+1]
            vectorizer = CountVectorizer()
            vectorizer.fit_transform(context)
            context_no_order = vectorizer.get_feature_names_out()
            cbows.append((val, context_no_order))
    return cbows

# Step 5: Call get_cbows() and store results
cbows = get_cbows(sentence_lst, context_length)

# Step 6: Define the skip-gram function (order preserved)
def get_skip_grams(sentence_lst, context_length):
    skip_grams = []
    for i, val in enumerate(sentence_lst):
        if i < context_length:
            pass
        elif i < len(sentence_lst) - context_length:
            context = sentence_lst[i-context_length:i] + sentence_lst[i+1:i+context_length+1]
            skip_grams.append((val, context))
    return skip_grams

# Step 7: Call get_skip_grams() and store results
skip_grams = get_skip_grams(sentence_lst, context_length)

# Step 8: Print CBOW results
try:
    print('\nContinuous Bag of Words')
    for cbow in cbows:
        print(cbow)
except:
    pass

# Step 9: Print skip-gram results
try:
    print('\nSkip Grams')
    for skip_gram in skip_grams:
        print(skip_gram)
except:
    pass
````

People use **CBOW** instead of **skip-grams** in several situations because CBOW has specific advantages that make it useful depending on the goals and constraints of a project.

Here are the main reasons:

---

## **1. CBOW is *much faster* to train**

CBOW predicts a target word from its surrounding context **as a bag of words**, meaning it ignores order.
Because it compresses all context words into a single representation, it requires **fewer computations** than skip-grams.

This makes CBOW ideal when:

* you have a **very large corpus**,
* you need **high-speed training**,
* or you are working with **limited computational resources**.

---

## **2. CBOW performs well for frequent words**

CBOW tends to produce better embeddings for:

* **common, high-frequency words**
* **general semantic similarity tasks**

Because frequent words have many training examples, CBOW aggregates context efficiently and produces stable representations.

---

## **3. CBOW is less sparse and more stable**

Skip-grams generate many (word → context) pairs.
CBOW reduces multiple context words into a single average representation, which:

* smooths out noise,
* reduces sparsity,
* often produces more stable embeddings in low-data settings.

---

## **4. CBOW is conceptually simpler**

CBOW is easier to:

* implement,
* explain,
* tune.

The context window is just a bag-of-words.

---

## **5. CBOW works well when order isn’t important**

For many semantic tasks, the **exact order of neighboring words doesn’t matter**.
CBOW captures the idea of *what kinds of words tend to appear near this word*, not *in what order*.

---

## **Why skip-grams are often preferred today**

Even though people still use CBOW, skip-grams are used more frequently in modern NLP because:

* they handle **rare words** better,
* they capture **richer patterns** (including word order),
* they produce **better-quality embeddings** for many downstream tasks.

---

## **Summary Table**

| Method         | Strengths                                                        | Weaknesses                                      |
| -------------- | ---------------------------------------------------------------- | ----------------------------------------------- |
| **CBOW**       | Fast, stable, good for frequent words, simple                    | Ignores order, poorer performance on rare words |
| **Skip-grams** | Better for rare words, captures order, higher-quality embeddings | Slower to train, more computationally expensive |

# Gensim

When we train word embeddings, the **corpus of text** we choose determines how the vectors are shaped. Different corpora lead to different embeddings because each corpus provides its own patterns of word usage.

* A **large, generic corpus** (e.g., news + web text) → more **generalizable** word embeddings.
* A **small, domain-specific corpus** (e.g., a single play or a TV series) → embeddings that capture **very specific usage and relationships** in that domain.

## Training Our Own Word2Vec Model with Gensim

So far, we’ve used **pre-trained word embeddings** from **spaCy**. Those models were trained (using word2vec) on large collections of blog posts and news articles from the Linguistic Data Consortium.

With **gensim**, we can instead train **our own word2vec model** on any corpus we like.

A word2vec model can be trained with:

```python
model = gensim.models.Word2Vec(
    corpus,
    size=100,
    window=5,
    min_count=1,
    workers=2,
    sg=1
)
```

Where:

* `corpus` – a **list of lists**, where each inner list is a document, and each element of that inner list is a **word token**.
* `size` – the **dimensionality** of the word embeddings (e.g., 100 dimensions).
* `window`, `min_count`, `workers`, `sg` – additional training parameters (you don’t need to worry about them in detail right now).

## Inspecting the Vocabulary

Once the model is trained, you can view the **vocabulary** it learned:

```python
vocabulary_of_model = list(model.wv.vocab.items())
```

This gives you all the words that appear in the training corpus often enough to be included in the model.

## Domain-Specific Meaning

When we train on a **smaller, specific corpus**, we capture the unique way words are used there.

Example:

* A general spaCy model might not place **“Ross”** and **“Rachel”** especially close together.
* A gensim model trained on **Friends** scripts would likely learn that “Ross” and “Rachel” occur in very similar contexts and place their vectors close together in the embedding space.

## Finding Similar Words with `.most_similar()`

To explore the learned relationships, we can ask gensim for the words most similar to a given term:

```python
model.most_similar("my_word_here", topn=100)
```

* `"my_word_here"` – the **target word**.
* `topn` – how many most similar words to return.

This tells us which words the model believes share similar contexts and thus similar meanings.

## Odd-One-Out with `.doesnt_match()`

Gensim also provides a neat utility to detect which word **doesn’t fit** with the others:

```python
model.doesnt_match(["asia", "mars", "pluto"])
```

Here, gensim returns the word whose vector is **furthest** from the others in the list — the **odd-one-out** according to the embedding space.

---

Using gensim, we’re no longer limited to generic, pre-trained embeddings. We can **shape word embeddings around our own corpus**, capturing the specific relationships, style, and semantics that matter for our particular task or dataset.

````python
# Step 1: Import gensim, NLTK stopwords, and the Romeo & Juliet text
import gensim
from nltk.corpus import stopwords
from utils2.romeo_juliet import romeo_and_juliet

# Step 2: Load English stop words
stop_words = stopwords.words('english')

# Step 3: Preprocess the text: lowercase, split, and remove stop words
romeo_and_juliet_processed = [
    [word for word in romeo_and_juliet.lower().split() if word not in stop_words]
]

# Step 4: Inspect the first 20 tokens of the inner list
print(romeo_and_juliet_processed[0][:20])

# Step 5: Train a gensim Word2Vec model (100-dimensional embeddings)
model = gensim.models.Word2Vec(
    romeo_and_juliet_processed,
    vector_size=100,
    window=5,
    min_count=1,
    workers=2,
    sg=1
)

# Step 6: Extract and print the vocabulary of the trained model
vocabulary = list(model.wv.key_to_index.items())  # (word, index)
print(vocabulary[:20])
````

````python
similar_to_romeo = model.wv.most_similar("romeo", topn = 20)
similar_to_romeo
````

````python
less_similar = model.wv.doesnt_match(["romeo", "juliet", "mercutio"])
less_similar
````

````python
less_similar = model.wv.doesnt_match(["cat", "dog", "mountain"])
less_similar
````

# Review

* **Vectors** are containers of numerical information and can range from 1 dimension to hundreds or thousands of dimensions.
* **Word embeddings** are vector representations of words, where words appearing in similar contexts are mapped to points that lie close together in vector space.
* **spaCy** allows us to load and use **pre-trained word embedding models**.
* **Distance metrics** help compare vectors; for high-dimensional embeddings, **cosine distance** is the most effective measure.
* **Word2Vec** is a shallow neural network that learns word embeddings using either **Continuous Bag-of-Words (CBOW)** or **Skip-grams**.
* **Gensim** provides tools to **train custom word embedding models** on any corpus we choose.


````python
# Step 1: Import spaCy and cosine distance
import spacy
from scipy.spatial.distance import cosine

# Step 2: Load the word embedding model
nlp = spacy.load("en_core_web_sm")

# Step 3: Create word embedding vectors
sponge_vec = nlp("sponge").vector
starfish_vec = nlp("starfish").vector
squid_vec = nlp("squid").vector

# Step 4: Compute cosine distances between vectors
dist_sponge_star = cosine(sponge_vec, starfish_vec)
dist_sponge_squid = cosine(sponge_vec, squid_vec)
dist_star_squid = cosine(starfish_vec, squid_vec)

# Step 5: Print distances
print(dist_sponge_star)
print(dist_sponge_squid)
print(dist_star_squid)
````
